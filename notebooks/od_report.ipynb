# Oscillation Detection — Report

```python
import sys
from pathlib import Path

_server = Path(SERVER_DIR) if SERVER_DIR else (Path.cwd().parent / "server").resolve()
sys.path.insert(0, str(_server))

def _resolve_run(run_dir, usecase):
    from postprocess import detect_usecase
    if run_dir:
        return Path(run_dir)
    runs = sorted((_server / "runs").glob("*/*"), key=lambda p: p.stat().st_mtime, reverse=True)
    for r in runs:
        try:
            if (r / "wiring.json").exists() and detect_usecase(r) == usecase:
                return r
        except Exception:
            continue
    raise RuntimeError(f"No {usecase.upper()} run found under {_server / 'runs'}")

RUN = _resolve_run(RUN_DIR, "od")
```
